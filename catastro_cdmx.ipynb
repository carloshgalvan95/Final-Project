```python
# Files to load
catastro_alvaro_obregon = "db_catastro/catastro_alvaro_obregon.csv"
catastro_azcapotzalco = "db_catastro/catastro_azcapotzalco.csv"
catastro_benito_juarez = "db_catastro/catastro_benito_juarez.csv"
catastro_coyoacan = "db_catastro/catastro_coyoacan.csv"
catastro_cuajimalpa = "db_catastro/catastro_cuajimalpa.csv"
catastro_cuauhtemoc = "db_catastro/catastro_cuauhtemoc.csv"
catastro_gustavo_a_madero = "db_catastro/catastro_gustavo_a_madero.csv"
catastro_iztacalco = "db_catastro/catastro_iztacalco.csv"
catastro_iztapalapa = "db_catastro/catastro_iztapalapa.csv"
catastro_magdalena_contreras = "db_catastro/catastro_magdalena_contreras.csv"
catastro_miguel_hidalgo = "db_catastro/catastro_miguel_hidalgo.csv"
catastro_milpa_alta = "db_catastro/catastro_milpa_alta.csv"
catastro_tlahuac = "db_catastro/catastro_tlahuac.csv"
catastro_tlalpan = "db_catastro/catastro_tlalpan.csv"
catastro_venustiano_carranza = "db_catastro/catastro_venustiano_carranza.csv"
catastro_xochimilco = "db_catastro/catastro_xochimilco.csv"
```
